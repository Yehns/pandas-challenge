```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load (Remember to Change These)
school_data_to_load = os.path.join("resources\\schools_complete.csv")
student_data_to_load = os.path.join("resources\\students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools (dexcluding the header)
total_schools = len(school_data_complete.school_name.unique())

# Calculate the total number of Students
total_students = (school_data_complete.student_name.count())

# Calculate the total budget
total_budget = school_data_complete.budget.sum() 

# Calculate the average maths score
aver_maths_score = school_data_complete.maths_score.mean()

# Calculate the average reading score
aver_reading_score = school_data_complete.reading_score.mean()

# Calculate the percentage of students with a passing maths score greater than 50
passing_math_50 = (school_data_complete[school_data_complete['maths_score']>50].maths_score.count())/total_students*100

# Calculate the percentage of students with a passing reading score greater than 50
passing_reading_50 = (school_data_complete[school_data_complete['reading_score']>50].reading_score.count())/total_students*100

# Calculate the percentage of students that passed maths and reading with a score greater than 50
pass_rate = (passing_math_50 + passing_reading_50)/2
pass_rate

# Create a dataframe to hold the above results
area_summary = pd.DataFrame({"Total Schools": [total_schools], 
                        "Total Students": [total_students],
                        "Total Budget": [total_budget],
                        "Average Maths Score": [aver_maths_score],
                        "Average Reading Score": [aver_reading_score],
                        "Percentage of Students Passing Maths": [passing_math_50],
                        "Percentage of Students Passing Reading":[passing_reading_50],
                        "Pass Maths and Reading": [pass_rate]})


# Reformat the display of information to be cleaner

area_summary['Total Students'] = area_summary.apply(lambda x: "{:,.0f}".format(x['Total Students']), axis=1)
area_summary['Total Budget'] = area_summary.apply(lambda x: "${:,.2f}".format(x['Total Budget']), axis=1)
area_summary['Average Maths Score'] = area_summary.apply(lambda x: "{:,.1f}".format(x['Average Maths Score']), axis =1)
area_summary['Average Reading Score'] = area_summary.apply(lambda x: "{:,.1f}".format(x['Average Reading Score']), axis =1)
area_summary['Percentage of Students Passing Maths'] = area_summary.apply(lambda x: "{:,.1f}".format(x['Percentage of Students Passing Maths']), axis =1)
area_summary['Percentage of Students Passing Reading'] = area_summary.apply(lambda x: "{:,.1f}".format(x['Percentage of Students Passing Reading']), axis =1)
area_summary['Pass Maths and Reading'] = area_summary.apply(lambda x: "{:,.1f}".format(x['Pass Maths and Reading']), axis =1)
area_summary
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name

  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Groupby (set the first columns) the "school_names" from the complete data set
school_summary = school_data_complete.groupby(['school_name'])

# Sort and diplay the school names 
school_names = school_data_complete.school_name.sort_values().unique()

# Sort school type values by school name on adjacent axis
school_types = school_data.sort_values(by="school_name").type

# Display the total students of each school thorugh count function
school_total_students = list(school_summary.student_name.count())

# Display school budget - as the data for school on row is the same, use the mean function to extract single value
school_budget = list(school_summary.budget.mean())

# Display per student budget using zip() function - set 'a' value (school budget) divided by 'b' (total students) in list function
school_per_student_budget = [a/b for a,b in zip(school_budget,school_total_students)]
# Display the average math score for each school
school_avg_math_score = list(school_summary.maths_score.mean())

# Display the average reading score for each school
school_avg_reading_score = list(school_summary.reading_score.mean())

# From compete data determie students passing maths (above 50%)
maths_pass = school_data_complete[school_data_complete['maths_score'] > 50].groupby(['school_name'])
    # Using the zip() function - set 'a'value (number of students from each school passing maths) divided by the 'b'value (total school students) to determine the percentage of students passing maths
students_passing_maths = [(a/b)*100 for a,b in zip(maths_pass.maths_score.count(),school_total_students)]
# From compete data determie students passing reading (above 50%)
reading_pass = school_data_complete[school_data_complete['reading_score'] > 50].groupby(['school_name'])
     # Using the zip() function - set 'a'value (number of students from eachschool passing reading) divided by the 'b'value (total school students) to determine the percentage of students passing reading
students_passing_reading= [(a/b)*100 for a,b in zip(reading_pass.reading_score.count(),school_total_students)]
# Determine the overall passing rate for students (passing both reading and maths) using the zip() function - set 'a' value (students passing maths) and dviided by the 'b' value (students passing reading) multiplied by 2
school_overall_passing = [(a+b)/2 for a,b in zip(students_passing_maths,students_passing_reading)]

# Create a dataframe to hold the results
per_school_summary = pd.DataFrame({"School Names":school_names,
                                  "School Type":school_types,
                                  "Total Students":school_total_students,
                                  "Total School Budget":school_budget,
                                  "Per Student Budget":school_per_student_budget,
                                  "Average Math Score":school_avg_math_score,
                                  "Average Reading Score":school_avg_reading_score,
                                  "% Passing Math":students_passing_maths,
                                  "% Passing Reading":students_passing_reading,
                                  "% Overall Passing":school_overall_passing})

# Reformat the data to improve readability
per_school_summary["Average Math Score"] = per_school_summary.apply(lambda x: "{:,.1f}".format(x["Average Math Score"]), axis =1)
per_school_summary["Average Reading Score"] = per_school_summary.apply(lambda x: "{:,.1f}".format(x["Average Reading Score"]), axis =1)
per_school_summary["% Passing Math"] = per_school_summary.apply(lambda x: "{:,.1f}".format(x["% Passing Math"]), axis =1)
per_school_summary["% Passing Reading"] = per_school_summary.apply(lambda x: "{:,.1f}".format(x["% Passing Reading"]), axis =1)
per_school_summary["% Overall Passing"] = per_school_summary.apply(lambda x: "{:,.1f}".format(x["% Overall Passing"]), axis =1)
per_school_summary['Total School Budget'] = per_school_summary.apply(lambda x: "${:,.2f}".format(x['Total School Budget']), axis=1)
per_school_summary['Per Student Budget'] = per_school_summary.apply(lambda x: "${:,.2f}".format(x['Per Student Budget']), axis=1)
per_school_summary


```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python

# For all years 9 to 12:
        # Use groupby to organise the average of maths scores for each year, in every school (creating an independant value for each school)
        # If the school data (school_data_complete) is equal to the set year (9,10,11,12) it will group each school (school_name) independantly and average all maths scores (maths_scores)
year_9 = school_data_complete.loc[school_data_complete.year == 9].groupby("school_name")["maths_score"].mean()
year_10= school_data_complete.loc[school_data_complete.year == 10].groupby("school_name")["maths_score"].mean()
year_11 = school_data_complete.loc[school_data_complete.year == 11].groupby("school_name")["maths_score"].mean()
year_12 = school_data_complete.loc[school_data_complete.year == 12].groupby("school_name")["maths_score"].mean()

# Print the dataframe, sorted by the school_name and each following column representing a year at school
maths_scores_by_year = pd.DataFrame({
        "Year 9": year_9,
        "Year 10": year_10,
        "Year 11": year_11,
        "Year 12": year_12
})

# Reformat the data to improve readability
maths_scores_by_year['Year 9'] = maths_scores_by_year['Year 9'].map('{:,.2f}'.format)
maths_scores_by_year['Year 10'] = maths_scores_by_year['Year 10'].map('{:,.2f}'.format)
maths_scores_by_year['Year 11'] = maths_scores_by_year['Year 11'].map('{:,.2f}'.format)
maths_scores_by_year['Year 12'] = maths_scores_by_year['Year 12'].map('{:,.2f}'.format)
maths_scores_by_year


```
